```python
import pandas as pd
```

```python
# create a datafram using list

names= ['Ayush','Abhinav','Amit']
pd.DataFrame(names)
```

```python
names= ['Ayush','Abhinav','Amit']
age = [25,30,35]
pd.DataFrame(names,age)
```

```python
names= ['Ayush','Abhinav','Amit']
age = [25,30,35]
pd.DataFrame(zip(names,age))
```

```python
names= ['Ayush','Abhinav','Amit']
age = [25,30,35]
city=['Blr','del','hyd']
data = zip(names,age)
pd.DataFrame(data,city)
```

```python
names= ['Ayush','Abhinav','Amit']
age = [25,30,35]
city=['Blr','del','hyd']
col = ['Name','Age','city']
data = zip(names,age,city)
pd.DataFrame(data,columns=col)
```

```python
#add new column
names= ['Ayush','Abhinav','Amit']
age = [25,30,35]
city=['Blr','del','hyd']

job=['DS','ML','AI']
col = ['Name','Age','city','job']

data = zip(names,age,city,job)
pd.DataFrame(data,columns=col)
```

```python
#update the existing column
names= ['Ayush','Abhinav','Amit']
age = [60,70,80]
city=['Blr','del','hyd']

job=['DS','ML','AI']
col = ['Name','Age','city','job']

data = zip(names,age,city,job)
pd.DataFrame(data,columns=col)
```

```python
#Change the index
names= ['Ayush','Abhinav','Amit']
age = [60,70,80]
city=['Blr','del','hyd']

data=zip(names,age,city)
col = ['Name','Age','city']
idx =['A','B','C']

df = pd.DataFrame(data,
                 index=idx,
                 columns=col)
df
```

```python
#shape
print("the number of rows are",df.shape[0])
print("the number of column are",df.shape[1])
```

```python
#Drop column
df
df.drop('city',axis=1)
df
```

```python
df.drop('C',axis=0)
df
```

```python
names=['Ayush','Avinash','Akash','Anvi']
age=[25,30,35,40]
city=['hyd','blr','chennai','Mumbai']
data=zip(names,age,city)
cols=['Names','Age','City']
idx=['A','B','C','D']
df=pd.DataFrame(data,
            index=idx,
                columns=cols)
df
```

```python
#remove rows
df.drop('C',axis=0,inplace=True)
df
```

```python
df.drop('D',axis=0)
```

```python

#Save data to csv or excel

names=['Ayush','Avinash','Akash','Anvi']
age=[25,30,35,40]
city=['hyd','blr','chennai','Mumbai']
data=zip(names,age,city)
cols=['Names','Age','City']
idx=['A','B','C','D']
df=pd.DataFrame(data,
            index=idx,
                columns=cols)
df
```

```python
df.to_csv("test.csv",index=False)
```
